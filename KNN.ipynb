```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
classified = pd.read_csv('Classified Data')
classified = classified.drop('Unnamed: 0',axis=1)
classified
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(classified.drop('TARGET CLASS', axis=1))
```

```python
scaled_values = scaler.transform(classified.drop('TARGET CLASS', axis=1))
```

```python
scaled_values
```

```python
df = pd.DataFrame(scaled_values, columns = classified.columns[:-1])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X = df
y = classified['TARGET CLASS']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
error_rate = []

for i in range(1,40):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
    
```

```python
plt.plot(range(1,40),error_rate)
plt.grid('on')
```

```python
knn= KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
